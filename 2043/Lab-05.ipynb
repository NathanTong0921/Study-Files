## K-th Largest Element

**Problem Description:**

 Given an integer array nums and an integer k, return the k-th largest element in the array. You must solve it without a full sort.

**Hint:**

**Forbidden Approach:** Sorting the entire array in descending order and returning the element at index k-1. This takes O(n log n) time, but the problem asks for a different or more optimized solution.

**Divide and Conquer Approach (Quickselect):**
This algorithm is an excellent application of the Quicksort idea and a pure divide and conquer algorithm. While Quicksort aims to sort the entire array, Quickselect's goal is only to find a single element, making it more efficient on average.

**The Divide-and-Conquer Trilogy:**

**1. Divide:** 

- Select an element from the array (preferably at random) to be the pivot.
- Call a partition function (the core operation in Quicksort) to rearrange the array. After partitioning, the pivot is placed at its final sorted position, pivot_index. All elements smaller than the pivot are on one side, and all larger elements are on the other.
- The array is now divided into two parts by the pivot.

**2. Conquer:**

- Now, we compare pivot_index with the index of the k-th largest element we're looking for (which is n-k in a 0-indexed array sorted smallest to largest).
- If pivot_index == n - k, then the pivot is exactly the element we want. Problem solved!
- If pivot_index < n - k, the k-th largest element is larger than the current pivot and must be in the right subarray. We only need to recursively search on the right side.
- If pivot_index > n - k, the k-th largest element is smaller than the current pivot and must be in the left subarray. We only need to recursively search on the left side.


**Time Complexity:**

**- Average Case:** Since we (ideally) halve the problem size at each step, the recurrence relation is T(n) = T(n/2) + O(n) (where O(n) is the cost of partitioning). This resolves to an average time complexity of O(n).

**- Worst Case:** If the chosen pivot is always the smallest or largest element, the problem size only decreases by 1, degrading the algorithm to O(n²). Randomizing pivot selection makes the worst case highly unlikely.


```python
import random

def partition(arr, low, high):
    """
    Partition function, same as in Quicksort.
    To avoid the worst case, a random pivot is chosen and swapped with the last element.
    """
    rand_pivot_index = random.randint(low, high)
    arr[rand_pivot_index], arr[high] = arr[high], arr[rand_pivot_index]
    
    pivot = arr[high]
    i = low - 1 
    
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
            
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quick_select(arr, low, high, k_smallest_index):
    """
    Recursively uses quickselect to find the k-th smallest element.
    """
    if low <= high:
        pivot_index = partition(arr, low, high)
        
        if pivot_index == k_smallest_index:
            return arr[pivot_index]
        elif pivot_index < k_smallest_index:
            return quick_select(arr, pivot_index + 1, high, k_smallest_index)
        else:
            return quick_select(arr, low, pivot_index - 1, k_smallest_index)

def find_kth_largest(nums, k):
    """
    Finds the k-th largest element in an array.
    """
    n = len(nums)
    # The k-th largest element is equivalent to the (n - k)-th smallest element (0-indexed).
    k_smallest_index = n - k
    
    # We operate on the array in-place to save space.
    return quick_select(nums, 0, n - 1, k_smallest_index)

# --- Test Cases ---
# Sample Input 1:
# n=6, k=2, nums=[10, 8, 2, 1, -10, 4] -> 2nd largest is 8
nums1 = [10, 8, 2, 1, -10, 4]
k1 = 2
# Pass a copy because the function modifies the list
print(f"In array {nums1}, the {k1}-th largest element is: {find_kth_largest(list(nums1), k1)}") # Expected: 8

# Sample Input 2:
# n=3, k=1, nums=[1, 4, 3] -> 1st largest is 4
nums2 = [1, 4, 3]
k2 = 1
print(f"In array {nums2}, the {k2}-th largest element is: {find_kth_largest(list(nums2), k2)}") # Expected: 4
```

## Max-Min Problem (3-Way Divide and Conquer)

**Problem Description:**

Design a 3-way split divide-and-conquer algorithm to find the maximum difference between any two elements in a given integer array A, which is max(A) - min(A). A 3-way split means you must divide the problem into three subproblems of sizes as equal as possible.

**Hint:**

**Standard Approach:** The simplest way is to iterate through the array once, keeping track of the max_val and min_val encountered. After the loop, max_val - min_val is the answer. This has a time complexity of O(n).

**Required Approach (3-Way Divide and Conquer):** This problem is pedagogical, designed to make you practice the divide-and-conquer mental model, even if it's not the most practical implementation for this specific task.

**The Divide-and-Conquer Trilogy:**

**1. Divide:** Split the array segment A[low...high] into three roughly equal subarrays:

- [low, mid1]
- [mid1 + 1, mid2]
- [mid2 + 1, high]
  
**2. Conquer:**

**Base Case:** When a sub-array is very small (e.g., 1 or 2 elements), we can directly find its max and min by comparison.
**Recursive Step:** Recursively call the function to find the max and min of each of the three sub-arrays.
- (max1, min1) = solve(A, low, mid1)
- (max2, min2) = solve(A, mid1 + 1, mid2)
- (max3, min3) = solve(A, mid2 + 1, high)

The maximum value of the original array [low...high] is the maximum of the three sub-array maximums: max(max1, max2, max3). Similarly, the minimum value is min(min1, min2, min3).


**Time Complexity:** The recurrence relation is T(n) = 3T(n/3) + O(1). According to the Master Theorem, the time complexity is O(n).

```python
def max_min_3way_divide_conquer(arr, low, high):
    # Base Case: If the segment has one or two elements.
    if high - low <= 1:
        if arr[low] > arr[high]:
            return arr[low], arr[high]
        else:
            return arr[high], arr[low]
            
    # Divide step
    mid1 = low + (high - low) // 3
    mid2 = low + 2 * (high - low) // 3
    
    # Conquer step (recursive calls)
    max1, min1 = max_min_3way_divide_conquer(arr, low, mid1)
    max2, min2 = max_min_3way_divide_conquer(arr, mid1 + 1, mid2)
    max3, min3 = max_min_3way_divide_conquer(arr, mid2 + 1, high)
    
    # Combine step
    final_max = max(max1, max2, max3)
    final_min = min(min1, min2, min3)
    
    return final_max, final_min

def find_max_difference(arr):
    max_val, min_val = max_min_3way_divide_conquer(arr, 0, len(arr) - 1)
    return max_val - min_val

# --- Test Cases ---
# Sample Input 1:
arr1 = [10, 2, 3, 4] # max=10, min=2, diff=8
print(f"The maximum difference for array {arr1} is: {find_max_difference(arr1)}") # Expected: 8

# Sample Input 2:
arr2 = [42]
print(f"The maximum difference for array {arr2} is: {find_max_difference(arr2)}") # Expected: 0

# Sample Input 3:
arr3 = [-5, -1, -3, 0, 7, 2] # max=7, min=-5, diff=12
print(f"The maximum difference for array {arr3} is: {find_max_difference(arr3)}") # Expected: 12
```

## Finding the Repeated Element

**Problem Description:**

You are given a sorted array A of n integers. This array contains all integers from 0 to n-2 exactly once, except for one integer which appears twice. You need to design a divide and conquer algorithm to find this repeated element.

**Hint:**

This problem is another excellent test of your ability to adapt binary search.

Key Insight:

- If the array were "perfect" (no duplicates), then for every index i, we would have A[i] == i.
  
- The presence of a duplicate breaks this perfect correspondence at some point.
  
- Suppose the repeated number is d. The array will look like: [0, 1, ..., d-1, d, d, d+1, ..., n-2].
  
- We can observe:
  - Before the second occurrence of the duplicate d, the condition A[i] >= i holds true.
  - Starting from the second occurrence of d, the array values "lag behind" their indices, meaning A[i] < i.
  - For example, if the array is [0, 1, 2, 2, 3] (n=5, d=2):
    - A[0]=0, A[1]=1, A[2]=2 (here A[i] == i)
    - A[3]=2 (here A[3] < 3). This is the first index where A[i] < i.
- Our goal is to find the first index i where A[i] < i. The value at this index, A[i], will be the duplicate number we are looking for.

**Binary Search Algorithm:**

- Initialize low = 0, high = n-1.
  
- In the loop, check the middle index mid:
  
    - If A[mid] < mid: This means the breaking point is at or to the left of mid. A[mid] is a candidate for our answer, but we must check the left half to see if an even earlier break exists. So, we record A[mid] as a potential answer and shrink our search space to [low, mid - 1].
  
    - If A[mid] >= mid: This means the segment from 0 to mid is "healthy". The breaking point must be to the right of mid. We should continue searching in [mid + 1, high].

When the loop terminates, the last candidate answer we recorded will be the correct one.

Time Complexity: This is a classic binary search, O(log n).

```python
def find_duplicate(arr):
    """
    In a sorted array containing numbers from 0 to n-2 with one duplicate, finds the duplicate number.
    """
    low = 0
    high = len(arr) - 1
    duplicate = -1
    
    while low <= high:
        mid = low + (high - low) // 2
        
        # If arr[mid] < mid, the break point is at or to the left of mid.
        # arr[mid] is a candidate. Record it and search the left for an earlier break.
        if arr[mid] < mid:
            duplicate = arr[mid]
            high = mid - 1
        # If arr[mid] >= mid, the 0..mid section is "healthy".
        # The break point must be to the right.
        else:
            low = mid + 1
            
    return duplicate

# --- Test Cases ---
# Sample Input 1:
arr1 = [0, 1, 1, 2, 3, 4] # n=6, contains 0-4, 1 is duplicated.
print(f"The duplicate element in array {arr1} is: {find_duplicate(arr1)}") # Expected: 1

# Sample Input 2:
arr2 = [0, 1, 2, 2] # n=4, contains 0-2, 2 is duplicated.
print(f"The duplicate element in array {arr2} is: {find_duplicate(arr2)}") # Expected: 2

# Sample Input 3:
arr3 = [0, 1, 2, 3, 3, 4] # n=6, contains 0-4, 3 is duplicated.
print(f"The duplicate element in array {arr3} is: {find_duplicate(arr3)}") # Expected: 3

# Boundary case
arr4 = [0, 0, 1, 2] # 0 is duplicated.
print(f"The duplicate element in array {arr4} is: {find_duplicate(arr4)}") # Expected: 0
```
