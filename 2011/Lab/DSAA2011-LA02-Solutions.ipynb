**Question 1.** Prove $1/\theta$ is the expectation of exponential distribution with
\begin{align}
f_X(x ; \theta)=\left\{\begin{array}{cc}
\theta \exp (-\theta x) & x \geq 0 \\
0 & x<0
\end{array}. \right.
\end{align}

**Solution**:

Step 1: definition of expectation
$$
\mathbb{E}[X] = \int_{-\infty}^{\infty} x f_X(x; \theta)  dx = \int_{0}^{\infty} x \cdot \theta e^{-\theta x}  dx
$$

Step 2: use integration by parts

Forumla for integration by parts:
$$
\int u  dv = uv - \int v  du
$$

$$
\begin{align*}
E[X] &= \theta \int_0^{\infty} x e^{-\theta x}dx \\
&= \theta \left[-\frac{x}{\theta}e^{-\theta x}\right]_0^{\infty} + \theta \int_0^{\infty} \frac{1}{\theta}e^{-\theta x}dx \\
&= \theta \left[0 - 0\right] + \int_0^{\infty} e^{-\theta x}dx \\
&= \left[-\frac{1}{\theta}e^{-\theta x}\right]_0^{\infty} \\
&= \frac{1}{\theta}
\end{align*}
$$



**Question 2.** Show that if we have $n$ vectors $\mathbf{x}_i  \in\mathbb{R}^d, i = 1,\ldots, n$ and $n$ scalars $a_i\in \mathbb{R}, i = 1,\ldots, n$, a linear function satisfies
$$
f\bigg(\sum_{i=1}^n a_i\  \mathbf{x}_i\bigg) =  \sum_{i=1}^n a_i\ f(\mathbf{x}_i) .
$$
Hint: begin with the case of $n=2$.

**Solution:**

We are tasked with proving that if we have $n$ vectors $x_i \in \mathbb{R}^d$ ($i = 1, \dots, n$)  
and $n$ scalars $a_i \in \mathbb{R}$, a linear function $f$ satisfies:
$$
f\left(\sum_{i=1}^n a_i x_i\right) = \sum_{i=1}^n a_i f(x_i).
$$

---

**Step 1: Start with the case $n = 2$.**

Using the linearity of $f$, we know that:
1. **Additivity**: $f(x + y) = f(x) + f(y)$ for all $x, y \in \mathbb{R}^d$,  
2. **Homogeneity**: $f(cx) = c f(x)$ for all $c \in \mathbb{R}$ and $x \in \mathbb{R}^d$.

For two vectors $x_1, x_2 \in \mathbb{R}^d$ and scalars $a_1, a_2 \in \mathbb{R}$, consider their linear combination:
$$
f(a_1 x_1 + a_2 x_2).
$$

Using the additivity property:
$$
f(a_1 x_1 + a_2 x_2) = f(a_1 x_1) + f(a_2 x_2).
$$

Now, using the homogeneity property:
$$
f(a_1 x_1) + f(a_2 x_2) = a_1 f(x_1) + a_2 f(x_2).
$$

Thus, for $n = 2$, we have shown:
$$
f(a_1 x_1 + a_2 x_2) = a_1 f(x_1) + a_2 f(x_2).
$$

---

**Step 2: Induction for general $n$.**

Assume that for $n = k$, the property holds:
$$
f\left(\sum_{i=1}^k a_i x_i\right) = \sum_{i=1}^k a_i f(x_i).
$$

Now, for $n = k + 1$, consider:
$$
f\left(\sum_{i=1}^{k+1} a_i x_i\right).
$$

Split the summation into two parts:
$$
f\left(\sum_{i=1}^{k+1} a_i x_i\right) = f\left(\sum_{i=1}^k a_i x_i + a_{k+1} x_{k+1}\right).
$$

Using additivity:
$$
f\left(\sum_{i=1}^k a_i x_i + a_{k+1} x_{k+1}\right) = f\left(\sum_{i=1}^k a_i x_i\right) + f(a_{k+1} x_{k+1}).
$$

By the induction hypothesis:
$$
f\left(\sum_{i=1}^k a_i x_i\right) = \sum_{i=1}^k a_i f(x_i).
$$

And by homogeneity:
$$
f(a_{k+1} x_{k+1}) = a_{k+1} f(x_{k+1}).
$$

Combining these results:
$$
f\left(\sum_{i=1}^{k+1} a_i x_i\right) = \sum_{i=1}^k a_i f(x_i) + a_{k+1} f(x_{k+1}).
$$

Thus:
$$
f\left(\sum_{i=1}^{k+1} a_i x_i\right) = \sum_{i=1}^{k+1} a_i f(x_i).
$$

By induction, the property holds for all $n$.

---

**Conclusion:**

We have shown that for any $n$ vectors $x_i \in \mathbb{R}^d$ and scalars $a_i \in \mathbb{R}$, a linear function $f$ satisfies:
$$
f\left(\sum_{i=1}^n a_i x_i\right) = \sum_{i=1}^n a_i f(x_i).
$$

**Question 3.** Let $f: \mathbb{R}\to\mathbb{R}$ be the absolute value function 
$$f(x) =|x|.$$
Is $f$ linear?

**Solution:**

A function $f : \mathbb{R} \to \mathbb{R}$ is linear if it satisfies **additivity** and **homogeneity of scalar multiplication** (if either property is not satisfied, we can say the function is not linear):

1. **Additivity**: $f(x + y) = f(x) + f(y)$ for all $x, y \in \mathbb{R}$.
2. **Homogeneity**: $f(c x) = c f(x)$ for all $x \in \mathbb{R}$ and $c \in \mathbb{R}$.

---

**Step 1: Check additivity**

Let $f(x) = |x|$. Consider two cases:

- Case 1: $x$ and $y$ have the same sign (e.g., both are positive or both are negative).  
  In this case, $|x + y| = |x| + |y|$, so additivity holds.
  
- Case 2: $x$ and $y$ have opposite signs.  
  For example, if $x > 0$ and $y < 0$, $|x + y| \neq |x| + |y|$ in general, because $|x + y|$ depends on the relative magnitudes of $x$ and $y$.

Thus, **additivity is not satisfied** for all $x, y \in \mathbb{R}$.

---

**Step 2: Check homogeneity**

Let $c \in \mathbb{R}$ and $f(x) = |x|$. Consider two cases:

- Case 1: $c \geq 0$.  
  $f(c x) = |c x| = c |x| = c f(x)$, so homogeneity holds.
  
- Case 2: $c < 0$.  
  $f(c x) = |c x| = |c| |x| = (-c) f(x) \neq c f(x)$, so homogeneity fails when $c < 0$.

---

**Conclusion:**

Since $f(x) = |x|$ does not satisfy additivity and fails homogeneity when $c < 0$, the function $f$ **is not linear**.

**2. Calculate the least squares solution using **only numpy**. Print the vector $\bar{\mathbf{w}}$.**

Suppose we have a linear model:

$$
y = X\beta + \varepsilon
$$

Where:
- $ y \in \mathbb{R}^{n \times 1} $: target values
- $ X \in \mathbb{R}^{n \times d} $: feature matrix
- $ \beta \in \mathbb{R}^{d \times 1} $: parameters to learn
- $ \varepsilon $: noise term

We aim to minimize the squared error:

$$
\min_\beta \|X\beta - y\|_2^2
$$

---

**Derivation**

Objective function:

$$
J(\beta) = \|X\beta - y\|_2^2 = (X\beta - y)^T(X\beta - y)
$$

Take gradient and set it to zero:

$$
\nabla_\beta J = 2X^T(X\beta - y) = 0
$$

This leads to the **normal equation**:

$$
X^T X \beta = X^T y
$$

If $ X^T X $ is invertible:

$$
\hat{\beta} = (X^T X)^{-1} X^T y
$$

---

```python
import numpy as np

# Example 1 data
X1 = np.array([[1, -7], [1, -5], [1, 1], [1, 5]])
y1 = np.array([-6, -4, -1, 4])

# Least squares solution
w1 = np.linalg.inv(X1.T @ X1) @ X1.T @ y1
print("w* for Example 1:", w1)


# Example 2 data
X2 = np.array([[1, 1, 1],
               [1, -1, 1],
               [1, 1, 3],
               [1, 1, 0]])
y2 = np.array([1, 0, 2, -1])

# Least squares solution
w2 = np.linalg.inv(X2.T @ X2) @ X2.T @ y2
print("w* for Example 2:", w2)
```

```python
# Generate 10 random samples and predict for example 1
x_new1 = np.random.uniform(-10, 10, size=(10, 1))
X_new1 = np.hstack([np.ones((10, 1)), x_new1])
y_pred1 = X_new1 @ w1
print("Predictions for Example 1:", y_pred1)

# Generate 10 random samples and predict for example 2
x_new2 = np.random.uniform(-5, 5, size=(10, 2))
X_new2 = np.hstack([np.ones((10, 1)), x_new2])  # Add bias term
y_pred2 = X_new2 @ w2
print("Predictions for Example 2:", y_pred2)
```

**Question 5.** 
In many real-world applications, we are not only interested in predicting a single outcome from the same set of features, but also multiple different yet related outcomes. For example:
* In e-commerce, the features (such as product price, ad spending, season) can be used to predict online sales and offline sales simultaneously.
* In healthcare, patient features (blood pressure, glucose level, heart rate) can be used to predict both disease risk and prognosis score.

This scenario is called multi-output (or multi-target) modeling, where we model a mapping from feature space to target space: $f: \, \mathbb R^d \to \mathbb R^c$.
* $d$ denotes **feature dimension**, i.e., the number of variables that describe each input sample. 
* $c$ denotes **number of output channels**, i.e., the number of results we want to predict for each input.

Consider fitting a general formed linear model: $\hat Y = XW + b^\top$, where $X \in \mathbb R^{n \times d}$, $Y \in \mathbb R^{n \times c}$ and $W \in \mathbb R^{d \times c}$, $b \in \mathbb R^{c}$ ($n$ denotes the total number of samples). Complete the following tasks by filling in all `TODO`s (using `numpy` only):

1. Synthetic data generation: We generate synthetic data by specifying true parameters and adding random noise to validate the performance of multi-output linear regression. We choose `np.random.default_rng` as the random number generator, you can use `rng.normal(..)` to generate synthetic data. We need to generate the ground truth parameters $(W_{true}, b_{true})$ and features $X$. Their components are drawn from normal distribution $\mathcal N(0, 1)$, for example, $W_{true}$= rng.normal(size=(d, c), loc=0, scale=1). 
   * Set $n = 300, \, d = 5 ,\, c = 2$
   * Generate $W_{true}$ and $b_{true}$
   * Build synthetic data: $Y = XW_{true} + b_{true}^\top + \varepsilon^\top$, where $\varepsilon_i \sim \mathcal N(0, \sigma^2), \sigma = 0.5$
2. Compute the least square solution for $(\hat W, \hat b)$
   * Find the least squares solution using **only** `np.linalg.solve` or `np.linalg.inv` (do not use `np.linalg.lstsq`).
   * Find the least squares solution using `np.linalg.lstsq`.
   * Compare the two results obtained above using `np.allclose'.
3. Prediction & Evaluation
   * Predict $\hat Y$ using $(\hat W, \hat b)$ and $X$.
   * For each output channel, compute the root mean square error (RMSE):
$$
\sqrt{\frac{1}{n}\sum_{i = 1}^{n} (y_{i, c} - \widehat {y_{i, c}})^2}
$$

```python
# ---------------------------
# 1) Data generation
# ---------------------------

def set_seed(seed: int):
    return np.random.default_rng(seed)

def generate_synthetic_data(rng, n=300, d=5, c=2, sigma=0.5):
    """
    Returns:
        X: (n, d)
        Y: (n, c)
        W_true: (d, c)
        b_true: (c,)
    """
    # Hints:
    # - X ~ N(0,1)
    # - W_true, b_true ~ N(0,1)
    # - noise ~ N(0, sigma^2)
    # - Y = X @ W_true + 1*b_true^T + noise
    X = rng.normal(size=(n, d))
    W_true = rng.normal(size=(d, c))
    b_true = rng.normal(size=(c,))
    noise = rng.normal(scale=sigma, size=(n, c))
    Y = X @ W_true + b_true + noise
    return X, Y, W_true, b_true


# ---------------------------
# 2) Model fitting
# ---------------------------

def add_bias_column(X):
    """
    Returns augmented design matrix A = [1, X].
    Shape: (n, d+1)
    """
    n = X.shape[0]
    return np.hstack([np.ones((n, 1)), X])

def fit_handcraft(X, Y):
    """
    Closed-form least square solution using normal equations on augmented features.
    Returns:
        W_hat: (d, c)
        b_hat: (c,)
    """
    A = add_bias_column(X)              # (n, d+1)
    ATA = A.T @ A                       # (d+1, d+1)
    ATY = A.T @ Y                       # (d+1, c)
    theta = np.linalg.solve(ATA, ATY)   # (d+1, c)
    b_hat = theta[0, :]                 # (c,)
    W_hat = theta[1:, :]                # (d, c)
    return W_hat, b_hat

def fit_lstsq(X, Y):
    """
    Closed-form least square solution using np.linalg.lstsq on augmented features.
    Returns:
        W_hat: (d, c)
        b_hat: (c,)
    """
    A = add_bias_column(X)                       # (n, d+1)
    theta, *_ = np.linalg.lstsq(A, Y, rcond=None)  # (d+1, c)
    b_hat = theta[0, :]                          # (c,)
    W_hat = theta[1:, :]                         # (d, c)
    return W_hat, b_hat


# ---------------------------
# 3) Prediction & Metrics
# ---------------------------

def predict(X, W, b):
    """
    Returns predictions: X @ W + 1*b^T
    Shape: (n, c)
    """
    n = X.shape[0]
    return X @ W + np.ones((n, 1)) @ b[None, :]

def rmse_per_output(Y_true, Y_pred):
    """
    Returns per-output RMSE as shape (c,)
    """
    mse = np.mean((Y_true - Y_pred) ** 2, axis=0)
    return np.sqrt(mse)

# ---------------------------
# Main script
# ---------------------------

rng = set_seed(911)
n, d, c = 300, 5, 2

# 1) Data
X, Y, W_true, b_true = generate_synthetic_data(rng, n=n, d=d, c=c, sigma=0.5)

# 2) Fit OLS (closed-form) and via lstsq
W_hc, b_hc = fit_handcraft(X, Y)
W_ls, b_ls = fit_lstsq(X, Y)

# 3) Cross-check solutions
# Use a tight tolerance; adjust if needed for numeric stability.
print("allclose(W_hc, W_ls):", np.allclose(W_hc, W_ls, atol=1e-8))
print("allclose(b_hc, b_ls):", np.allclose(b_hc, b_ls, atol=1e-8))

# 4) Train predictions & RMSE
Y_pred = predict(X, W_hc, b_hc)
rmse = rmse_per_output(Y, Y_pred)
print("Train RMSE per output:", rmse)  # shape (c,)
```
